```python
cols=['hr', 'holiday', 'weekday', 'weathersit', 'temp',
                              'atemp', 'hum', 'windspeed', 'casual', 'registered', 'cnt']
final_dataset[cols]=final_dataset[cols].apply(pd.to_numeric)
```

```python
final_dataset
```

```python
def find_outliers(df,cols):
    q1=df[cols].quantile(.25)
    q3=df[cols].quantile(.75)
    iqr=q3-q1
    min_n=q1-1.5*iqr
    max_n=q3+1.5*iqr
    outliers_index=df.index[(df[cols]<   min_n) | (df[cols]<   max_n)]
    return outliers_index
```

```python
find_outliers(final_dataset,'holiday')
```
